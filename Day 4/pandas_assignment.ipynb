```python
#Read titanic dataset and perform following operations and observe output
import pandas as pd
import numpy as np
read_df = pd.read_excel('/content/Titanic-Dataset (2).xlsx')
read_df

#2. Find max and min fairprice of each Pclass (1 and 2 and 3)
#3. Handle missing values in age and cabin columns
#4. Find average fair price of lower class people (Pclass = 3) who survived
#5. Learn dna implement how to find duplicate columns and rows and how to remove them
```

```python
#2. Find max and min fairprice of each Pclass (1 and 2 and 3)
Pclass_stats = read_df.groupby('Pclass')['Fare'].agg(['max','min'])
Pclass_stats
```

```python
#. 3Handle missing values in age and cabin columns
df_no_missing = read_df.copy()
df_no_missing['Age'] = df_no_missing['Age'].fillna(df_no_missing['Age'].median())
df_no_missing['Cabin'] = df_no_missing['Cabin'].fillna('Unknown')
df_no_missing
```

```python
#4. Find average fair price of lower class people (Pclass = 3) who survived

average_fare = read_df[(read_df['Pclass'] == 3) & (read_df['Survived'] == 1)]['Fare'].mean()
print(f"The average fare for lower class survivors is: {average_fare}")
```

```python
#5. Learn dna implement how to find duplicate columns and rows and how to remove them
duplicate_cols = read_df.columns[read_df.columns.duplicated()]
print("Dupliacte columns:", duplicate_cols)
read_df = read_df.loc[:, ~read_df.columns.duplicated()]
duplicate_rows = read_df[read_df.duplicated()]
print("\nDuplicate rows:\n", duplicate_rows)
read_df = read_df.drop_duplicates()
print("\nDataFrame after removing duplicates:\n", read_df)
```
